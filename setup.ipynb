# Development Environment 

## Python 
The Python interpretor and tools are all bundled in Python for Windows, available at the official [Python](https://www.python.org/downloads/) site. i'm not familiar with the new *Python install manager*, so for now ignore the main download button and use the link below it to download **standalone installer for Python 3.14.6**. 

I already have it installed so can't quickly see what the setup is like, so just accept defaults, and if asked to install you you or all users, choose all users, as sometimes you will want to run code as admin. 

To check the installation, open Windows Terminal and type `py`. This should open the Interactive Shell, or REPL. This is good  for running simple, single-line statements, and although a lot more can be done, it quickly gets painful, and it's hardly veer necessary. Type `print("Hello world")` + ENTER and it prints `Hello world`. Type `quit()` or `exit()` to exit the REPL. In modern Python `quit` and `exit` also work because these are shortcuts to the actual functions. It really doesn't matter, because you don't use these functions in normal code. 

Your brief intro to the REPL should look something like this: 

```py
Python 3.14.6 (tags/v3.14.6:c63aec6, Jun 10 2026, 10:26:10) [MSC v.1944 64 bit (AMD64)] on win32
Type "help", "copyright", "credits" or "license" for more information.
>>> print("Hello world")
Hello world
>>> quit 
```

## Visual Studio Code
VScode, as opposed to full-on Visual Studio, is by far the best editor I have ever used for Python, even compared with full scale commercial IDEs like JetBrains' PyCharm. It's totally free, and you can use it for countless other things, even non-coding editing. 

Most commands like creating, opening, and saving files are very simple and easy to access, and the other ones should become familiar quite quickly. 

Follow the *System setup* guide on the official [VSCode site](https://code.visualstudio.com/docs/setup/windows) to get started. This is better than the *User setup* because sometimes you might sometimes want run code as admin. 

You can Open VSCode from the start menu, but this opens it in your user folder and you have to navigate to your working folder. I find it easier to find the parent of your working folder in Explorer, right-click your working folder and choose **Open with Code**. You can also navigate to the working folder in the terminal and just enter the command `code`. 

To confirm that Python is available in VSCode, open a built-in terminal with the menu command `Terminal->New Terminal`, or the keyboard shortcut ``Ctrl+` ``. That's Ctrl and backtick. Then type `py` to enter a REPL, and `quit` to exit.
