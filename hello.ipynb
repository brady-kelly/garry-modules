## Output and Input
Python is most commonly used for scripting, AI and machine learning, and web development, and seldom for GUI work, so for a good while, all of our I/O will be to and from console.

### In the beginning was *Hello, World*.
It's inevitable, so lets get it out of the way.  Output to console is via the [print](https://docs.python.org/3/library/functions.html#print) function.
The simplest use of this function is to take an object, a string literal here, and write it to the console. Try and make this call output *Hello, world!*:

```python
print("wrong string")
```

### Getting Input
Input from the console is via the [input](https://docs.python.org/3/library/functions.html#input) function. Its simplest use is without any parameters, and in simply reads a string value from console and returns that value, which we assign to a variable, but it's more commonly used with the `prompt` paramter to indicate what input is wanted:

```python
name = input("What is your name?")
name
```

In the previous cell you can see two concepts:
- I didn't need to use [print](https://docs.python.org/3/library/functions.html#print) to show the value that was entered. When the interpretor sees an *expression*, in this case just the variable name, it evaluates it and prints its value.
- The `name` variable. All that is needed to declare a variable in Python is to assign a value to it, and it's type is automatically determined by the interpretor based on the value assigned to it. In this case, [input](https://docs.python.org/3/library/functions.html#input) returns a string, so name has a string type of that value.

### Numeric Input
Because [input](https://docs.python.org/3/library/functions.html#input) always returns a `string`, to get say an integer, you need to call the [int](https://docs.python.org/3/library/functions.html#int) function with that string: 

```python
age = int(input("How old are you?"))
age
```

Python is relatively simple with only three [standard numeric types](https://docs.python.org/3/library/stdtypes.html#typesnumeric):
- `int`, for integers, obtained from expressions that evaluate to integer, e.g: integer literals or from the [int](https://docs.python.org/3/library/functions.html#int) function.
- `float`, for floating point numbers, obtained from expressions that evaluate to floating point values or from the [float](https://docs.python.org/3/library/functions.html#float) function. 
- `complex`, for complex numbers, with real and imaginary parts, obtained from expressions that evaluate to complex number values or from the [complex](https://docs.python.org/3/library/functions.html#complex) function.

For now I will just focus on integers to simplify examples that aren't focused on numeric processing.
