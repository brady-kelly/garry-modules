## Python Classes
Python combines the three *object oriented*, *functional*, and *procedural* programming paradigms, and includes OOP's [class](https://docs.python.org/3/tutorial/classes.html#a-first-look-at-classes) construct as a means of encapsulating data and functionality.

Classes are quite complex behind the scenes, but the magic of Python does a wondeful job of protecting you from it all, and it also does a lot for you, so you can just start using this mechanism without getting lost in too much detail. So, please bear with me here, as the magic will be more visible when you start using classes properly, a little bit later.

### Class Definitions
A Python class defintion is a compound statement with a suite of nested statements that define instance methods.

Because Python is dynamic and interpreted, instance attributes can only be defined after a class is instantiated. Once a class is instantiated, the special attribute `self` represents that instance, much like the `this` variable in other languages. However, although it is implicitly initialized when an instance is created, and implicitly passed in method calls, it must still always be explicitly defined as a paramter for methods. 

```python
# Define the class.
class RgbColour:
    
    def setRgb(self, r, g, b):
        self.red = r
        self.green = g
        self.blue = b
        
    def printRgb(self):
        print ("R:", self.red, "G:", self.green, "B:", self.blue)
        
        
# Create an instance.
purp = RgbColour()

# Set RGB values for purple.
purp.setRgb(128, 0, 128)

# Confirm the values were set.
purp.printRgb()
```

Notice how, in the example above, we define the class with the `setRgb` instance that method creates the three colour instance attributes of `red`, `green` and `blue` by assigning values to these attributes on the special `self` attribute, and notice how we don't have to pass this when calling that method. 

Because this pattern of creating an instance and setting attributes on it is so common, Python wraps it up for us in a  special method named `__init__`. Python uses this naming convention of double underscores on each side of a name to indicate special functions that are called for you by the interpretor. These are known as *dunder* functions. The `__init__` function is automatically called when we instantiate a class, like a constructor in other languages, and it can take any parameters needed to set initial values on an instance.

 We can rewrite the `RgbColour` class as follows:

```python
# Define the class.
class RgbColour:
    
    def __init__(self, r, g, b) -> None:
        self.red = r
        self.green = g
        self.blue = b
                    
    def printRgb(self):
        print ("R:", self.red, "G:", self.green, "B:", self.blue)
    
        
# Create an instance with RGB values for purple.
purp = RgbColour(128, 0, 128)

# Confirm the values were set.
purp.printRgb()
```

Now the creation of the required instance attributes is automatic and coupled to the creation of all instances, improving the encapsulation of the `RgbColour` class.

### Inheritence
As in all OOP languages, Python classes support inheritence, and the basic syntax for a derived class is as follows:

```python
# Define the base class
class RgbColour:
    
    def __init__(self, r, g, b) -> None:
        self.red = r
        self.green = g
        self.blue = b
                    
    def printRgb(self):
        print ("R:", self.red, "G:", self.green, "B:", self.blue)

# Derive from RgbColour
class HexRgbColour(RgbColour):
        
    def __init__(self, r, g, b) -> None:
        super().__init__(r, g, b)
    
    def printHex(self):
        print(f"#{self.red:02X}{self.green:02X}{self.blue:02X}" )
        
# Create an instance.
purp = HexRgbColour(128, 0, 128)

# Print the colour in hex format.
purp.printHex()
```
