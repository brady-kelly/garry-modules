## Python Classes
Python is a mixture of the *object oriented*, *functional*, and *procedural* programming paradigms, and thus includes the [class](https://docs.python.org/3/tutorial/classes.html#a-first-look-at-classes) construct as a means of encapsulating data and functionality.

Classes are defined to have attributes (variables), and methods (functions) which are public by default. Methods are also virtual by default, meaning they can be overridden in derived classes.


### Class Definitions
A Python class defintion is a compount statement with a suite of nested statements that define attributes and methods:

```python
class RgbColour:
    colourType = "rgb" 

def __init__(self, name):
    self.colourName = name
```

Please bear with me here, because although this is quite a mouthful and seems complex, Python does a wondeful job of protecting you from it all, and just doing all the things for you when you start using these mechanisms.

When the above class definition occurs in code, it is executed, creating a *class object* (not an instance), with a `colourType` class variable named `colourType` and a special method named `__init__`, which is called whenever an `RgbColour` object is instantiated. This is essentially the same as a *constructor* in other OO languages. It is worth noting that classes can be instaniated without explicitly calling this method.

Data, or instance, attributes are defined by assigning values to variables on the instance, via the `self` attribute. This is a special attribute that must be explicitly defined on methods to provide access to the instance attrbutes and methods. Although it must always be explicitly defined, it is always implcitly passed to methods.
